<h3>Data Overview </h3>
<p> The below description is copied over from the official webpage for this dataset on lacity.gov <br>
This dataset reflects incidents of crime in the City of Los Angeles dating back to 2020. This data is transcribed from original crime reports that are typed on paper and therefore there may be some inaccuracies within the data. Some location fields with missing data are noted as (0°, 0°). Address fields are only provided to the nearest hundred block in order to maintain privacy. This data is as accurate as the data in the database. </p>

We have no convenient data given to us that we can map over all the crime codes to get the crime names. We will have to manually create a dictionary to map over the crime codes to get the crime names. <br>
Note that certain crime codes are not present in the document attached. We assume that these codes are meant for 'Other Crime'.

```python
def crime(i):
    homicide = ['homicide', 110, 113]
    rape = ['rape', 121, 122, 815, 820, 821]
    robbery = ['robbery', 210, 220]
    agg_assault = ['agg_assault', 230, 231, 235]
    domestic_violence = ['domestive_violence', 626, 627, 647, 763, 928, 930, 236, 250, 251, 761, 926]
    simple_assault = ['simple_assault', 435, 436, 437, 622, 623, 624, 625]
    burglary = ['burglary', 310, 320]
    mvt = ['mvt', 510, 520, 433]
    btfv = ['btfv', 330, 331, 410, 420, 421]
    personal_theft = ['personal_theft', 350, 351, 352, 353, 450, 451, 452, 453]
    lst = [homicide, rape, robbery, agg_assault, domestic_violence, simple_assault, burglary, mvt, btfv, personal_theft]
    for j in lst:
        if i in j:
            return j[0]
    else: return 'other_crime'
```

```python
dfc = df.copy()
dfc['Crime'] = dfc['Crm Cd'].map(crime)
```

```python
dfc = dfc.dropna(subset=['Crime'])
dfc.Crime
len(dfc)
```

```python
plt.figure(figsize=(11, 10))
cplot = sns.countplot(data=dfc, x='Crime', order=dfc['Crime'].value_counts().index)
plt.xticks(rotation=30, ha='right');
plt.xlabel('Crime Commited')
cplot = cplot.get_figure()
cplot.savefig('dist_by_crime.png')
```
